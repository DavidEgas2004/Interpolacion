#Interpolacion
Encontrar la ecuacion que pase por los puntos que mas se ajuste a esta, se asume cierto comportamiento de la curva en base a los puntos que se presenten.
Devuelve una funcion (polinomio) con la que se puede seguir hallando puntos.
##Series de taylor
Ayudan a realizar la interpolacion allrededor de un punto en base a la exactitud que se quiera obtener mediante el teorema de weistras.



##Graficar la serie taylor realizadas en la clase:

##- Funcion Cos(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5*np.pi, 5*np.pi, 500)
#limites de inicio y fin con un total de 500 pts

#Define la ecuacion de coseno actual
cos_x = np.cos(x)

taylor_2 = 1 - (x**2)/2
taylor_4 = 1 - (x**2)/2 + (x**4)/24
taylor_6 = 1 - (x**2)/2 + (x**4)/24 - (x**6)/720

plt.figure(figsize=(15, 10))
plt.plot(x, cos_x, label='cos(x)', color='red')
plt.plot(x, taylor_2, label='Segundo Orden', linestyle='--', color='blue')
plt.plot(x, taylor_4, label='Cuarto Orden', linestyle='--', color='green')
plt.plot(x, taylor_6, label='Sexto orden', linestyle='--', color='purple')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Serie de taylor de la funcion Cos(X))')
plt.legend()
plt.grid(True)
plt.xlim(-5, 5)
plt.ylim(-1.5, 1.5)
plt.show()
```

##- Funcion ln(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 10, 100)

#Ecuacion
ln=np.log(x)

taylor_1= (x-1) #Primer orden
taylor_2 = (x - 1) - ((x - 1) ** 2) / 2       # Segundo orden
taylor_3 = (x - 1) - ((x - 1) ** 2) / 2 + ((x - 1) ** 3) / 3  # Tercer orden

plt.figure(figsize=(15, 10))
plt.plot(x, ln, label='ln (x)', color='red')
plt.plot(x, taylor_1, label='Primer Orden', linestyle='--', color='blue')
plt.plot(x, taylor_2, label='Segundo Orden', linestyle='--', color='orange')
plt.plot(x, taylor_3, label='Tercer Orden', linestyle='--', color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Serie de taylor de la funcion ln(X))')
plt.legend()
plt.grid(True)
plt.xlim(0, 4)
plt.ylim(-2, 2)
plt.show()

```

##- Funcion $\frac{1}{1 - x}$

```python
import numpy as np
import matplotlib.pyplot as pltimport

# Definir el rango de x (evitar valores cercanos a 1 para evitar división por cero)
x = np.linspace(-0.9, 0.9, 100)

# Función 1/(1 - x)
f = 1 / (1 - x)

#taylor_1 = 1                        # Primer orden: 1 se la pone pero no se la grafica
taylor_2 = 1 + x                     # Segundo orden: 1 + x
taylor_3 = 1 + x + x**2              # Tercer orden: 1 + x + x^2
taylor_4 = 1 + x + x**2 + x**3       # Cuarto orden: 1 + x + x^2 + x^3
taylor_5 = 1 + x + x**2 + x**3 + x**4 # Quinto orden: 1 + x + x^2 + x^3 + x^4


plt.figure(figsize=(15, 10))
plt.plot(x, f, label=r'$\frac{1}{1 - x}$', color='blue')
plt.plot(x, taylor_2, label='Segundo Orden', linestyle='--', color='green')
plt.plot(x, taylor_3, label='Tercer Orden', linestyle='--', color='purple')
plt.plot(x, taylor_4, label='Cuarto Orden', linestyle='--', color='orange')
plt.plot(x, taylor_5, label='Quinto Orden', linestyle='--', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Serie de Taylor de la función 1/(1 - x)')
plt.legend()
plt.grid(True)
plt.xlim(-0.9, 0.6)
plt.ylim(-1, 2)
plt.show()
```
